```python
import re
mask = df_orig['Date'].map(lambda x: re.search('(03|04|05|06)/.+/2018', x) != None)
mask
```

```python
df = df_orig[mask]
df
```

```python
df['Full Time']=pd.to_datetime(df['Date'] + ' ' + df['Time'], format = '%m/%d/%Y %H:%M:%S')
```

```python
#df_dev = df[(df['SCP'] == '01-03-00')&(df.Unit == 'R010') & (df.Station =='42 ST-PORT AUTH')].sort_values('Full Time')
df_dev = df[(df['SCP'] == '01-03-00')&(df['C/A'] == 'N060') & (df.Station =='42 ST-PORT AUTH')].sort_values('Full Time')

df_dev
```

```python
gdf = df.groupby(['Station','SCP','C/A']).Entries.agg(['min','max'])
```

```python
gdf['Total Entries'] = gdf['max'] - gdf['min']
ggdf = pd.DataFrame(gdf.groupby('Station')['Total Entries'].sum()).sort_values('Total Entries',ascending=False)
ggdf_10= ggdf.head(10)
```

```python
ggdf_10['Total Entries'].plot.barh()
plt.show()
```
